# Textual Horizontal Bar Diagram — Specification (for Jupyter Notebook)

## Task
Write a program that reads numbers from a file and prints a textual horizontal bar diagram.

## Input
The program must:
- Read the file name from standard input.
- Open and read the file line by line.

Each line in the file may contain:
- A non-negative integer  
- Or an invalid value (non-integer text, empty line, negative number)

## Processing Rules
- Each line in the file corresponds to one numbered output row.
- Number rows starting from 1.
- If the line contains a valid non-negative integer `n`:
  - Print `n` stars separated by a single space.
  - Format:
    ```
    <row_number>. * * * ...
    ```
- If the line contains an invalid value:
  - Print only:
    ```
    <row_number>. 
    ```
    (no stars after the dot, just a space)
- Count only valid stars toward the total.
- After processing all lines, print:
    ```
    In total X stars.
    ```
    
where `X` is the total number of stars printed.

## Important Rules
- Strip whitespace from each file line.
- Negative numbers are treated as invalid.
- No extra blank lines before or after output.
- Exact spacing must match examples:
- One space after the dot.
- Stars separated by single spaces.

````python
try:
    filename = input().strip()
    file_opening = open(filename, 'r', encoding='utf-8')
except EOFError:
    raise SystemExit

# From task: number rows starting from 1
row_number = 1

# From task: count the total number of stars printed
total_stars = 0

for raw_line in file_opening:
    # From task: strip whitespace from each file line
    line = raw_line.strip()
    
    # flag for valid lines and counter for stars
    valid = False
    n = 0
    
    # From task: check if the line is not empty
    if line != '':
        try:
            # convert the line to an integer
            n = int(line)
            
            # From task: check if the integer is non-negative
            if n >= 0:
                valid = True
        except ValueError:
            valid = False 

    if valid:
        # From task: if the line contains a valid non-negative integer n
        if n > 0:
            # From task: print n stars separated by a single space
            stars = "* " * (n - 1) + "*"
            print(f"{row_number}. {stars}")
        else:
            # From task: if the line contains an invalid value
            print(f"{row_number}. ")
        
        # From task: count only valid stars toward the total
        total_stars += n
    else:
        # From task: if the line contains an invalid value
        print(f"{row_number}. ")

    row_number += 1

print(f"In total {total_stars} stars.")
````
