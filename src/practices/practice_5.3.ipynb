**Write a program that:**

Reads four lines from standard input in this order:

First name  
Last name  
E-mail  
Occupation  

Prints a formatted business card inside a frame.

**Formatting Rules**  
Combine first name, last name and e-mail into one line:

`FirstName LastName - email`

The frame width must be **38 characters inside** (not counting `|`).

The top and bottom borders must be:
+--------------------------------------+
 
(38 hyphens between the plus signs)

The content must be centered inside 38 characters using spaces.

The layout must be exactly:

+--------------------------------------+
|                                      |
|<centered name + email line>          |
|                                      |
|<centered occupation line>            |
|                                      |
+--------------------------------------+
 
 
There must be no extra blank lines before or after output.

Input must be stripped using .strip().



Here is an example of the program output.

 Please enter your first name: Malle
 Please enter your last name: Mets
 Please enter your e-mail: malle@ut.ee
 Please enter your occupation: Teacher

 +------------------------------------+
 |                                    |
 |      Malle Mets - malle@ut.ee      |
 |                                    |
 |              Teacher               |
 |                                    |
 +------------------------------------+
Hint:


>>> "-" * 20
------------------
>>> "Teacher".center(30)
Teacher 

```python
user_first_name = input("Please enter your first name: ")
user_last_name = input("Please enter your last name: ")
user_email = input("Please enter your e-mail: ")
user_occupation = input("Please enter your occupation: ")

user_data = f"{user_first_name} {user_last_name} - {user_email}"

# frame settings
width = 38
border = "+" + ("-" * width) + "+"

centered_name = user_data.center(width)
centered_occupation = user_occupation.center(width)

print(border)
print("|" + " " * width + "|")
print("|" + centered_name + "|")
print("|" + " " * width + "|")
print("|" + centered_occupation + "|")
print("|" + " " * width + "|")
print(border)
```
