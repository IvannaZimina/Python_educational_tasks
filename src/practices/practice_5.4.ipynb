**Task Description**
Write a program that converts cents into euros and cents.

The program must:

Read one line from standard input.  
Interpret it as a non-negative integer (number of cents).  
Convert the amount into euros and cents:

1 euro = 100 cents  
Euros = cents // 100  
Remaining cents = cents % 100  

**Output Rules**  
If both euros and cents are non-zero:

X euro(s) and Y cent(s)

If only euros are non-zero:

X euro(s)

If only cents are non-zero:

Y cent(s)

If the input is 0, print:

0 cents

**Singular / Plural Rules**  
Use *euro* when the amount is exactly 1, otherwise *euros*.  
Use *cent* when the amount is exactly 1, otherwise *cents*.

**Error Handling**  
If the input:  
– is not an integer  
– or is negative  

Print exactly:

Enter a valid amount!

No extra blank lines are allowed before or after output.  
Input must be stripped.

```python
number_of_cents = input().strip()

if not number_of_cents.isdigit():
    print("Enter a valid amount!")
else:
    cents = int(number_of_cents)

    euros = cents // 100 
    remaining = cents % 100

    if euros == 0 and remaining == 0:
        print("0 cents")
    else:
        parts = []

        if euros > 0:
            if euros == 1:
                euro_word = "euro"
            else:
                euro_word = "euros"

            euro_part = str(euros) + " " + euro_word
            parts.append(euro_part)

        if remaining > 0:
            if remaining == 1:
                cent_word = "cent"
            else:
                cent_word = "cents"

            cent_part = str(remaining) + " " + cent_word
            parts.append(cent_part)

        result = " and ".join(parts)

        print(result)
```
