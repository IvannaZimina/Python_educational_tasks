**Write a program that determines a personâ€™s gender from an Estonian Personal Identification Number (isikukood).**

The format of the Estonian Personal Identification Number is:

`GYYMMDDSSSC`

The first digit **G** indicates gender.

- If **G is odd**, the person is **male**.
- If **G is even**, the person is **female**.

The program must:

- Read one line from standard input (the personal identification number).
- Strip surrounding whitespace.
- Validate that:
  - The code consists of **exactly 11 digits**.
  - All characters are **numeric**.
  - The first digit is **between 1 and 8** (valid Estonian codes).

**Output Rules**

If valid:

- Print **Male** if first digit is odd.
- Print **Female** if first digit is even.

If invalid:

- Print **Enter a valid personal identification number!**

No extra blank lines are allowed before or after output.

```python
isikukood = input().strip()


if len(isikukood) == 11 and isikukood.isdigit() and isikukood[0] in "12345678":
    
    first_digit = int(isikukood[0])
    
    if first_digit % 2 == 0:
        print("Female")
    else:
        print("Male")

else:
    print("Enter a valid personal identification number!")
```
