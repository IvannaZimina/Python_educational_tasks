# Task: Uppercase Word Processing

## Write a program that:

- Reads one word from standard input.
- Converts the word to **uppercase**.
- Prints the uppercase word repeatedly so that the number of printed lines equals the **length of the word**.
- Each line must contain **exactly** the uppercase word.
- There must be **no extra blank lines** before or after the output.

---

## Input
One line containing a word (string).

---

## Output
- Print the uppercase word exactly **len(word)** times, each on its own line.
- If the input is an **empty line**, the program must **print nothing**.
- Whitespace around the word must be ignored (use `.strip()`).

``

```python
# 1. Reads one word from standard input.
word = input().strip()

# if the word isn't empty
if word:
    # 2.Converts the word to uppercase.
    upper_word = word.upper()
    
    # 3. Prints the uppercase word repeatedly so that the number of printed lines equals the length of the word.
    for _ in range(len(word)):
        print(upper_word)
```
