**Write a program that extracts a surname from a University of Tartu staff URL.**

Each staff member has a URL in the format:

https://oigus.ut.ee/et/tootaja/username

The username part consists of:

firstname-surname

The program must:

- Read one line from standard input (the URL).  
- Strip surrounding whitespace.  
- Extract the part after the last `/`.  
- Extract the surname (the part after the last `-`).  
- Print:

Surname is <surname>

The surname must be printed exactly as it appears in the URL (lowercase).

**Error Handling**  
If:

- The URL does not contain `/tootaja/`  
- OR there is no `-` in the username part  
- OR the surname part is empty  

Print exactly:

Enter a valid url!

No extra blank lines are allowed before or after output.

```python
url_input = input().strip()
url_parts = url_input.split("/")

if "tootaja" not in url_parts:
    print("Enter a valid url!")
    exit()

username = url_parts[-1]
surname = username.split("-")[-1]

if username == "" or "-" not in username or surname == "":
    print("Enter a valid url!")
    exit()

print(f"Surname is {surname}")
```
