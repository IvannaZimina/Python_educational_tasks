## Task: Print File Lines with Numbers (with Error Handling)

Write a program that:
1. Reads a **file name** from standard input.
2. Opens the file and prints **all its lines with line numbers**.
3. Handles the situation where the file does not exist using **try–except**.

### Requirements
- Each line must be printed in the format:

`<line_number>. <line_content>`

- Line numbering **starts from 1**.
- Preserve the **original line content** (without the trailing newline).
- Strip **only** the newline character (`\n`), not other spaces.
- If the file does not exist, print exactly:

`File not found!`

- **No extra blank lines** before or after output.
- **Do not** print input prompts in the output.

```python
# from task: read a file name from standard input
filename = input().strip()

# from task: handle missing file using try–except
try:
    # from task: open the file
    with open(filename, encoding="utf-8") as f:  
        # from task: line numbering starts from 1
        line_number = 1

        for line in f:
            # from task: strip only the newline character (\n), not other spaces
            content = line.rstrip('\n')

            # from task: each line must be printed in the format <line_number>. <line_content>
            print(f"{line_number}. {content}")   

            # increment numbering
            line_number += 1                     

# from task: print exactly this text when the file does not exist
except FileNotFoundError:
    print("File not found!")
```
