**Write a program that processes a studentâ€™s name and grades.**

The program must:

- Read the first name (string).
- Read the grades (string).
- Format and print the results as specified.

**Processing Rules**  
- Strip whitespace from both inputs.  
- Format the name so that:  
  - First letter is uppercase  
  - Remaining letters are lowercase  
- Convert all grades to uppercase.  
- Assume the number of grades is at least 2.

**Output Format**  
The program must print exactly the following lines (with exact wording):

Hello <FormattedName>, your grades are <UPPERCASE_GRADES>  
You have <number_of_grades> grades  
Your grade for the second course is <second_grade>  
The number of A's and B's is <count_of_A_and_B>

Where:

- `<number_of_grades>` is the total length of the grades string.  
- `<second_grade>` is the second character (index 1).  
- `<count_of_A_and_B>` counts both 'A' and 'B'.  

No extra blank lines before or after output.


```python
first_name = input().strip()
grades = input().strip()

first_name_formated = first_name[0].upper() + first_name[1:].lower()
grades_formatted = grades.upper()

if len(grades_formatted) < 2:
    print("Enter at least 2 grades!")
else:
    number_of_grades = len(grades_formatted)
    second_grade = grades_formatted[1]
    count_of_A_and_B = grades_formatted.count('A') + grades_formatted.count('B')

    print(f"Hello {first_name_formated}, your grades are {grades_formatted}")
    print(f"You have {number_of_grades} grades")
    print(f"Your grade for the second course is {second_grade}")
    print(f"The number of A's and B's is {count_of_A_and_B}")


```
