## Task: Read University of Tartu Staff URLs and Extract Surnames

Write a program that reads University of Tartu staff URLs from a file whose name depends on user input.

### The program must:
1. Read one integer **n** from standard input.  
2. Open the file:

``urls<n>.txt``

For example:
- If input is `1` → open `urls1.txt`
- If input is `5` → open `urls5.txt`

Each line in the file contains one URL in the format:

``https://oigus.ut.ee/et/tootaja/firstname-surname``

### The program must:
1. Open and read the file `urls<n>.txt`.  
2. For each line:
   - Strip surrounding whitespace.
   - Extract the username part (after the last `/`).
   - Extract the surname.

3. Print each surname on a separate line.

---

### Important Rules

You must implement a function:

```python
def find_name(url: str) -> str
```

The function must:

- Take a URL as a parameter.
- Return the surname (the part after the first `-`).
- If there are multiple dashes, everything after the first dash belongs to the surname.
  - Example:  
    `chen-wan-lin` → `wan-lin`

Assume all URLs in the file are valid.

No extra blank lines before or after output.

````python
from typing import List

# from task: implement required function
def find_name(url: str) -> str:
    url = url.strip()                      # from task: strip surrounding whitespace
    last_segment = url.rsplit('/', 1)[-1]  # from task: extract the username part (after the last '/')
    dash_pos = last_segment.find('-')      # from task: find the surname (the part after the first '-')
    
    # checking        
    if dash_pos == -1:
        surname = last_segment
    else:
        # from task: everything after first dash is the surname - like .slice[:]
        surname = last_segment[dash_pos + 1:]
    return surname

# from task: read one integer n from standard input
n_str = input().strip()                   
n = int(n_str)

# from task: open the file: urls{n}.txt
filename = f"urls{n}.txt"                 

# if the file didn't exist, the code will raise FileNotFoundError
try:
    # from task: open and read the file urls{n}.txt
    with open(filename, encoding="utf-8") as f:      
        for line in f:
            # from task: strip surrounding whitespace
            url = line.strip()                        
            if not url:
                continue

            # from task: use the required function
            surname = find_name(url)       

            # from task: print each surname on a separate line
            print(surname)
except FileNotFoundError:
    # optional (not from task): аккуратное сообщение вместо падения
    print(f"File not found: {filename}. Put the file next to the notebook or enter another n.")
````
